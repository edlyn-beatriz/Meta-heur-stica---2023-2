```python
#QUESTÃO 7
print("Digite 10 números positivos:\n")
cont = 0
sum = 0
for x in range(1,11):
    n = int(input())
    if n > 0:
        cont += 1
        sum += n
    sum /= cont
    print(sum)
```

```python
#QUESTÃO 8
x = float(input(print("Digite 10 números, informaremos o maior e o menor valor digitado\n")))
max = x
min = x
for x in range(1,11):
    x = float(input())
    if x > max:
        max = x
    if x < min:
        min = x

print("maior valor:",max, "\nmenor valor:", min)
```
